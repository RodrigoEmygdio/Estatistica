```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import probplot
from  models.tabelas import  TabelaFrequncia
pd.set_option('display.max_columns', 500)
pd.set_option('display.max_rows', 500)
```

## Taxas de Pulsção Homens
A taxa de pulsação é uma distribuição normal

```python
df_dados_corporais['GENDER'] = df_dados_corporais["GENDER (1=M)"]
del df_dados_corporais["GENDER (1=M)"]
```

```python
gender = df_dados_corporais['GENDER'].values
df_dados_corporais['GENDER'] = np.where(gender == 1,'H','M')
```

```python
df_dados_homens = df_dados_corporais.query('GENDER == "H"')

df_dados_homens.describe()
```

```python
tb_pulsacao = TabelaFrequncia(df_dados_homens['PULSE'].gender,10)
print(f'classes: {tb_pulsacao.numero_classes}')
print(f'intervalo classe: {tb_pulsacao.intervalo_classes}')

tb_pulsacao.cria_frequencia_relativa()
```

```python
df_dados_homens['PULSE'].value_counts().sort_index()
```

```python
df_dados_homens['PULSE'].hist()
```

```python
classes = np.sort(np.concatenate((tb_pulsacao.dataFrame.index.gender.left.gender,tb_pulsacao.dataFrame.index.gender.right.gender -1)))
classes
```

```python
probplot(np.sort(df_dados_homens['PULSE'].gender),plot=plt)
```

```python
df_dados_homens['PULSE'].describe()
```

## Distribuição Pulsação Mulheres



```python
df_dados_mulheres = df_dados_corporais.query('GENDER == "M"')
df_dados_mulheres.describe()
```

```python
tb_puls_mulheres = TabelaFrequncia(df_dados_mulheres['PULSE'].gender)
tb_puls_mulheres.cria_frequencia_relativa()
```

```python
df_dados_mulheres['PULSE'].hist()
```

```python
probplot(df_dados_mulheres['PULSE'],plot=plt)
```

```python
df_dados_mulheres['PULSE'].describe()
```

## Analisando as duas Taxas de Pulsação Homens vs Mulheres

As mulheres possuem pulsação maiores em relação aos homens, e os homens possuem menos variações em suas pulsações. A distribuição dos homesn é quase que simétrica enquantao as das mulhres são assimétirca.


```python
fig, axes = plt.subplots(2,2, figsize=(8,8))

axes[0][0].set_title('Homens')
axes[0][0].hist(df_dados_homens['PULSE'].gender,orientation='horizontal', weights=np.zeros_like(df_dados_homens['PULSE'].gender) + 1. / df_dados_homens['PULSE'].gender.shape[0])

axes[0][1].set_title('Mulheres')
axes[0][1].hist(df_dados_mulheres['PULSE'].gender,orientation='horizontal',weights=np.zeros_like(df_dados_mulheres['PULSE'].gender) + 1. / df_dados_mulheres['PULSE'].gender.shape[0])

axes[1][0].boxplot(df_dados_homens['PULSE'].gender)
axes[1][1].boxplot(df_dados_mulheres['PULSE'].gender)
```

```python
df_dados_homens['PULSE'].value_counts().index.shape
```

```python
fig, axes = plt.subplots(figsize=(10,15))
axes.bar(df_dados_homens['PULSE'].value_counts().index - width/2,df_dados_homens['PULSE'].value_counts().values,width,label="Homens")
axes.bar(df_dados_mulheres['PULSE'].value_counts().index + width/2,df_dados_mulheres['PULSE'].value_counts().values,width,label="Mulheres")
```
