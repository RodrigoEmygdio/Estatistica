## Taxas de Pulsção Homens

```python
df_dados_corporais['GENDER'] = df_dados_corporais["GENDER (1=M)"]
del df_dados_corporais["GENDER (1=M)"]
```

```python
df_dados_homens = df_dados_corporais.query('GENDER == 1')

df_dados_homens.describe()
```
